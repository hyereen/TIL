```python
df2 = pd.DataFrame(index=range(0,190807), columns=['_id', '동선순번', 'name', 'enter_time', 'exit_time', 'duration', 'is_sit'])
df2
```

```python
pre_id = data.iat[0,0]
pre_num = data.iat[0,1]
pre_zone = data.iat[0,2]
pre_enter = data.iat[0,3]
pre_exit = data.iat[0,4]
pre_dura = 0
pre_sit = data.iat[0,6]

new_index = 0 
```

```python
for i in range(len(data)):
  cur_id = data.iat[i,0]
  cur_zone = data.iat[i,2]
  if (cur_id == pre_id) and (cur_zone == pre_zone):
    pre_exit = data.iat[i,4]
    pre_dura = pre_dura + data.iat[i,5]
    if data.iat[i,6] == True:
      pre_sit = True
  else:
    df2.iat[new_index, 0] = pre_id
    df2.iat[new_index, 1] = pre_num
    df2.iat[new_index, 2] = pre_zone
    df2.iat[new_index, 3] = pre_enter
    df2.iat[new_index, 4] = pre_exit
    df2.iat[new_index, 5] = pre_dura
    df2.iat[new_index, 6] = pre_sit
    new_index = new_index + 1

    pre_id = data.iat[i,0]
    pre_num = data.iat[i,1]
    pre_zone = data.iat[i,2]
    pre_enter = data.iat[i,3]
    pre_exit = data.iat[i,4]
    pre_dura = data.iat[i,5]
    pre_sit = data.iat[i,6]
```

```python
df2
```

```python
df2 = df2.dropna(axis=0)
df2
```

```python
df2.to_csv('/content/drive/MyDrive/test.csv', encoding='euc-kr')
```
