# 결측치

```python
train.info()
```

```python
train_mass = train[['Body Mass (g)']]
train_mass
```

```python
train_for_cls = train.drop(['Body Mass (g)'], axis=1)
train_for_cls
```

```python
test_for_cls = test.copy()
```

```python
concat_data = pd.concat([train_for_cls, test_for_cls], axis=0).reset_index()
del concat_data['index']
concat_data
```

```python
concat_data.info()
```

```python
concat_data['Delta 15 N (o/oo)'].fillna(concat_data['Delta 15 N (o/oo)'].median(), inplace=True)
```

```python
concat_data['Delta 13 C (o/oo)'].fillna(concat_data['Delta 13 C (o/oo)'].median(), inplace=True)
```

```python
concat_data.info()
```

```python
concat_data
```

```python
train = concat_data[concat_data['Sex'].isnull()==False]
test = concat_data[concat_data['Sex'].isnull()==True]
```

```python
X_train = train.drop(['id', 'Sex'], axis=1)
Y_train = train[['Sex']]

X_test = test.drop(['id', 'Sex'], axis=1)
Y_test = test[['Sex']]
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder

# 라벨 인코더 생성
encoder = LabelEncoder()

# X_train데이터를 이용 피팅하고 라벨숫자로 변환한다
X_train['Species'] = encoder.fit_transform(X_train['Species'])
X_train['Island'] = encoder.fit_transform(X_train['Island'])
X_train['Clutch Completion'] = encoder.fit_transform(X_train['Clutch Completion'])
X_train
```

```python
X_test
```

```python
X_test['Species'] = encoder.fit_transform(X_test['Species'])
X_test['Island'] = encoder.fit_transform(X_test['Island'])
X_test['Clutch Completion'] = encoder.fit_transform(X_test['Clutch Completion'])
X_test
```

## RF로 성별 예측해서 결측치 채우기

```python
random_forest = RandomForestClassifier(n_estimators=100)

random_forest.fit(X_train, Y_train)

Y_pred = random_forest.predict(X_test)

Y_pred
```

```python
y_pred_idx = 0
for i in range(len(concat_data)):
    if pd.isna(concat_data.iloc[i].Sex):
        concat_data.iloc[i, 7] = Y_pred[y_pred_idx]
        y_pred_idx +=1
```
