```python
# iris 데이터 불러오기
import pandas as pd
import numpy as np

iris = pd.read_csv('./iris.csv')
```

```python
iris
```

```python
iris.target.unique()
```

```python
x = iris.iloc[:, 0:4]
y = iris.iloc[:, -1]
```

```python
# 전처리
from sklearn.preprocessing import LabelEncoder, RobustScaler

label = LabelEncoder()
y = label.fit_transform(y)

sc = RobustScaler()
x = sc.fit_transform(x)
```

```python
# train, test set 나누기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=99, stratify=y)

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
print(cross_val_score(RandomForestClassifier(), x_train, y_train, cv=3).mean())
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
predict = model.predict_proba(x_test)
output=pd.DataFrame({'setosa':predict[:,0], 'versicolor':predict[:,1], 'virginica':predict[:,2]})
#output.to_csv('123.csv', index=False)
```

```python
# 모델 평가
import sklearn.metrics as met
print(met.confusion_matrix(y_test, model.predict(x_test)))
print(met.accuracy_score(y_test, model.predict(x_test)))
```
